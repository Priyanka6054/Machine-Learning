## Data Preprocessing:

```python
# Combine train and test datasets for preprocessing

combined_data = pd.concat([train_data, test_data], axis=0, ignore_index=True)
```

```python
combined_data.head()
```

```python
# Identify numerical and categorical columns

numeric_cols = combined_data.select_dtypes(include=np.number).columns.tolist()
categorical_cols = combined_data.select_dtypes(exclude=np.number).columns.tolist()
```

```python
# Check the missing value

train_data.isnull().sum()
```

```python
# Impute missing values for numerical columns using KNNImputer

from sklearn.impute import KNNImputer
imputer = KNNImputer()
imputer
```

```python
combined_data[numeric_cols] = imputer.fit_transform(combined_data[numeric_cols])
```

```python
# One-hot encode categorical columns

combined_data = pd.get_dummies(combined_data, columns=categorical_cols)
```

## Feature Selection/Engineering:

```python
# Split the combined dataset back into train and test datasets

train_processed = combined_data[:len(train_data)]
test_processed = combined_data[len(train_data):]
```

```python
# Split the train dataset into features and target variable

X_train = train_processed.drop('SalePrice', axis=1)
y_train = train_processed['SalePrice']
```

```python
# Scale the features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler
```

```python
X_train_scaled = scaler.fit_transform(X_train)
```

```python
X_train_scaled.shape
```

##  Model Selection:

```python
# Initialize the model

from sklearn.ensemble import HistGradientBoostingRegressor
model = HistGradientBoostingRegressor()
model
```

```python
from sklearn.model_selection import KFold, cross_val_score
```

```python
# Define cross-validation strategy
kfold = KFold(n_splits=5, shuffle=True, random_state=42)
kfold
```

```python
# Perform cross-validation on the training set
cv_scores = cross_val_score(model, X_train_scaled, y_train, cv=kfold, scoring='neg_mean_squared_error')
cv_scores
```

```python
# Convert scores to positive
cv_scores = -cv_scores
```

```python
# Print mean and standard deviation of cross-validation scores

print("Mean RMSE:", np.mean(np.sqrt(cv_scores)))
print("Standard Deviation of RMSE:", np.std(np.sqrt(cv_scores)))
```

```python
# Train the model on the entire train dataset
model.fit(X_train_scaled, y_train)
```

```python
# Make predictions on the test dataset
X_test_scaled = scaler.transform(test_processed.drop('SalePrice', axis=1))
```

```python
predictions = model.predict(X_test_scaled)
```

```python
# Creating a DataFrame for the predictions
House_Price_Prediction = pd.DataFrame({'Id': test_data['Id'], 'SalePrice': predictions})
House_Price_Prediction
```
